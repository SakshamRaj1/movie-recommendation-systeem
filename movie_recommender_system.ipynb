```python
# The above output is a string of list of dictionay
# ['Adventure, 'Fantasy', Action', 'SciFi']
# We need to conver the data into this format.
```

```python
# convert into list of strings

import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]

# It is a String data type
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())

# The movies['overview'].apply(lambda x:x.split()) line. The split() method, when applied to a string, breaks that string into a list of substrings (words)
# based on whitespace by default. By using .apply(), I'm telling pandas to run this split() operation on each individual string in the 'overview' column.

movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

#**Text Vectorisation - using Bag of Words**

```python
# Convert text to vectors, the core idea is that the most similar movies vectors will be nearest to each other.
```

```python
# The Bag of Words (BoW) is a foundational Natural Language Processing (NLP) method that converts unstructured text into numerical data for machine learning algorithms.
# It ignores grammar and word order, determining importance entirely by how frequently words appear in a document/feature.
# We combine all the fetaures of all rows into a single cell and then take max_fetures = 5000 oe some other value.
# Then calculate the frequency of all the tags of a movie with respect to max_features.
# So, our data has 4806 movies, so each movie will have a vector of each feature. Hence, each movie will have 5000 vectors, which will create a 4806*5000 dimensional space.
# Stop words are common, high-frequency words (like "the", "and", "is", "in") filtered out during text processing and search engine indexing. They carry very little semantic meaning.

# We will use Scikit Learn Library for Text vectorisation, and then find the
```

# **Similarity Score - using Cosine Score**

This is becasue we are focussed on calculating the angle between the vectors (Cosine Distance) and not on their tip-to-tip distance (Euclidean Distance) for accurately measuring the similarity.

Euclidean Distance fails for high dimensional data.

The smaller the angle in cosine distance, the more it is similar.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)

# SImilarity Matrix
```

```python
similarity[0].shape
```

```python
similarity[1]
```

```python
# Logic building

new_df['title'] == 'Avatar'
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    # tells the index of the movie based on the movie name
    distances = similarity[movie_index]
    # tells the distances between movie vectors of the give movie index
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    '''sorts the distance in descending order and selects the 5 most closest (similar) movie, it gives the output of the form
     [(index, distance1), (index, distance2), (index, distance3), (index, distance4), (index, distance5)], eg., [(1, 1.0000), (814, 0.20500), (519, 0.18500), (2311, 0.15960),(4804, 0.11583)
    '''
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        # prints the title of the movies based on the index while fetching it from the new_df
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```
